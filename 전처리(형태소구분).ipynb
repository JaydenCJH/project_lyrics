```python
import pandas as pd
import os, re 
```

```python
txt_path = 'trot_text.txt'
raw_corpus = []

with open(txt_path, 'r', encoding = 'utf-8') as f:
    raw = f.read().splitlines()
    raw_corpus.extend(raw)
print(len(raw_corpus))
raw_corpus
```

```python
def preprocess_sentence(sentence):
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence)
    sentence = re.sub(r'[" "]+', " ", sentence)
    sentence = re.sub(r"[^a-zA-Z가-힣?.!,¿]+", " ", sentence)
    sentence = sentence.strip()
    sentence = re.sub(r"\(.\)", " ", sentence) # 7
    sentence = '<start> ' + sentence + ' <end>'
    return sentence
```

```python
corpus = []
for sentence in raw_corpus:
    if len(sentence) == 0:
        continue
    if sentence[-1] == ':':
        continue
    if len(sentence)>150:
        continue
    
    preprocessed_sentence = preprocess_sentence(sentence)
    corpus.append(preprocessed_sentence)
    
print(len(corpus))
print(corpus[:10])
```

```python
def list_to_string(words):

    in_tag = False
    result = []
    tag_content = ""
    
    for word in words:
        if word == '<':
            in_tag = True
            tag_content = word
        elif word == '>':
            in_tag = False
            tag_content += word
            tag_content = tag_content.replace(" ", "")
            result.append(tag_content)
            tag_content = ""
        elif in_tag:
            tag_content += ' ' + word
        else:
            result.append(word)

    result_string = ' '.join(result)
    return result_string
```

```python
from eunjeon import Mecab
mecab = Mecab()
cc = mecab.morphs(corpus[0])

list_to_string(cc)
```

```python
from eunjeon import Mecab
mecab = Mecab()

corpus_f = []
for i in range(0, len(corpus)):

  cc = mecab.morphs(corpus[i])

  corpus_f.append(list_to_string(cc))
  
corpus_f


```

```python
df = pd.DataFrame({'col':corpus_f})
df.to_csv('corpus_f.csv')
```
